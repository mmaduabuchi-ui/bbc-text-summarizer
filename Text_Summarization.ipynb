```python
import pandas as pd

# Load dataset (update path if different)
df = pd.read_csv("/content/bbc_news.csv")

# Show first 5 rows
df.head()
```

```python
# Select only title and description
news_df = df[['title', 'description']].dropna()

# Show first 5 rows
news_df.head()
```

```python
from transformers import pipeline

# Load a pre-trained summarization model (BART)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
# Summarize first 5 rows
for i in range(5):
    text = news_df['description'].iloc[i]

    summary = summarizer(text, max_length=50, min_length=10, do_sample=False)

    print(f"\n--- Article {i+1} ---")
    print("Original:", text)
    print("Summary:", summary[0]['summary_text'])
```

```python
# Limit to a smaller sample first (to avoid long runtime on CPU)
sample_df = news_df.head(50).copy()

# Generate summaries
summaries = []
for text in sample_df['description']:
    summary = summarizer(text, max_length=50, min_length=10, do_sample=False)
    summaries.append(summary[0]['summary_text'])

# Add new column
sample_df['summary'] = summaries

# Show result
sample_df.head()
```

```python
from tqdm import tqdm

# Step 7 (improved): Apply summarizer to full dataset with progress bar
summaries = []

for text in tqdm(news_df['description'], desc="Summarizing"):
    try:
        # Adjust max_length dynamically (about 40% of input words)
        words = text.split()
        max_len = max(10, int(len(words) * 0.4))   # at least 10 tokens
        min_len = min(5, max_len - 1)              # smaller than max_len

        summary = summarizer(
            text,
            max_length=max_len,
            min_length=min_len,
            do_sample=False
        )
        summaries.append(summary[0]['summary_text'])
    except Exception as e:
        summaries.append("Error: " + str(e))

# Add summaries to DataFrame
news_df['summary'] = summaries

# Save results
news_df.to_csv("bbc_summaries.csv", index=False)
print(" Summaries saved to bbc_summaries.csv")
```

```python
# Step 8a (fixed version) - Inspect random samples with better summary length
import random

# Pick 3 random samples from the dataset
sample_indices = random.sample(range(len(news_df)), 3)

for idx in sample_indices:
    text = news_df['description'].iloc[idx]

    summary = summarizer(
        text,
        max_length=120,   # increased from 50 to 120
        min_length=30,    # increased from 10 to 30
        do_sample=False
    )

    print(f"\n--- SAMPLE {idx} ---")
    print("Original:", text)
    print("Summary:", summary[0]['summary_text'])
```

```python
# Step 8b - Dynamic summarization length
import random

sample_indices = random.sample(range(len(news_df)), 3)

for idx in sample_indices:
    text = news_df['description'].iloc[idx]

    input_len = len(text.split())  # word count
    # max_length is capped to be slightly less than input length
    max_len = min(120, max(20, int(input_len * 0.8)))
    min_len = max(10, int(max_len * 0.3))

    summary = summarizer(
        text,
        max_length=max_len,
        min_length=min_len,
        do_sample=False
    )

    print(f"\n--- SAMPLE {idx} ---")
    print("Original:", text)
    print("Summary:", summary[0]['summary_text'])
```

```python
# Step 8c - Skip very short inputs
for idx in sample_indices:
    text = news_df['description'].iloc[idx]

    input_len = len(text.split())

    if input_len < 25:
        # too short, keep as-is
        summary_text = text
    else:
        max_len = min(120, max(20, int(input_len * 0.8)))
        min_len = max(10, int(max_len * 0.3))

        summary = summarizer(
            text,
            max_length=max_len,
            min_length=min_len,
            do_sample=False
        )
        summary_text = summary[0]['summary_text']

    print(f"\n--- SAMPLE {idx} ---")
    print("Original:", text)
    print("Summary:", summary_text)
```
